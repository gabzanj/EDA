**EDA (Exploratory Data Analysis) using Sweetviz**
(From [Data Professor](https:/https://www.youtube.com/watch?v=UR_OK8vBpeY/))

```python
import pandas as pd
penguins = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/penguins_cleaned.csv')
penguins.columns
```

```python
# X: independent variable
# y: dependent variable

X = penguins.drop('species', axis=1)
y = penguins['species']
```

```python
X
```

Analyze the Dataset (penguins)

```python
pip install sweetviz
```

```python
import sweetviz as sv

analyze_report = sv.analyze(penguins)
analyze_report.show_html('analyze_penguins_df.html', open_browser=False)
```

```python
import IPython

IPython.display.HTML('analyze_penguins_df.html')
```

Side-by-Side Comparation of Train vc Test Set

Data Split using 80/20 Split Ratio

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Comparison report

```python
compare_report = sv.compare([X_train, 'Train'], [X_test, 'Test'])
```

```python
compare_report.show_html('compare.html', open_browser=False)
IPython.display.HTML('compare.html')
```
